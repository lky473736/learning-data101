```python
# 원 면적 구하기
from math import pi # 패키지 (내장되어 있는 함수들이 담겨짐)

R = int(input()) # input을 받을 때는 무조건 문자열 형태로 받음
S = R * R * pi
print (S)
```

```python
# 원 면적 구하는 함수
def get_size(R) : # 여기서의 변수는 parameter
    return R * R * pi

S = get_size(10) # 여기서의 실수값은 argument
print (S)
```

```python
# 터틀 그래픽으로 정n각형 그리기
import turtle
t = turtle.Turtle() # turtle 모듈에 있는 Turtle 클래스 객체를 t로 설정

N = int(turtle.textinput("angle?", "number of angle?"))
length = int(turtle.textinput("length?", "length?"))

for _ in range (N) :
    t.forward (length)
    t.left (360/N)
```
