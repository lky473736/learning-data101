```python
#  window 3  
# y = 2x, 3x, 4x

import math

x = [x for x in range (-50, 51)]
y1 = [2*compo for compo in x]
y2 = [3*compo for compo in x]
y3 = [4*compo for compo in x]

# plt.plot(x, y1, 'r-', x, y2, 'g-', x, y3, 'b-')
plt.plot(x, y1, label='2x')
plt.plot(x, y2, label='3x')
plt.plot(x, y3, label='4x')

plt.legend() #   (   label  )
plt.show()
```

```python
# , ,  

import math

x = list(range(-360, 361))
sine = [math.sin(math.radians(x)) for x in range (-360, 361)]
cosine = [math.cos(math.radians(x)) for x in range (-360, 361)]
tangent = []
for i in x :
    if i % 180 == 90 :
        tangent.append(float('nan'))  # 
    else :
        tangent.append(math.tan(math.radians(i)))

# print (sine, cosine, tangent)

plt.plot(x, sine, label='sine')
plt.plot(x, cosine, label='cosine')
plt.plot(x, tangent, label='tangent')
plt.ylim(-10, 10)  #   tangent 

plt.legend()
plt.show()
```

```python
#   == plt.bar()
#     

years = [1980 + 10 * i for i in range (1, 11)]
daramji = [200 + 2**i for i in range (0, 10)]

plt.bar(years, daramji, width=3)
plt.title('daramji')

plt.xticks (years, years) # x 

plt.show()
```

```python
#   

years = [1980 + 10 * i for i in range (1, 11)]
daramji = [200 + 2**i for i in range (0, 10)]
mouse = [80 + 1.5**i for i in range (0, 10)]
matbatji = [100 + 2.1**i for i in range (0, 10)]

plt.bar (years, daramji, label='daramji', width=3)
plt.bar (years, mouse, label='mouse', width=3)
plt.bar (years, matbatji, label='matbatji', width=3)

plt.show()
```

```python
#      (  )
#         ( )
# ndarray  

import numpy as np

# years = [1980 + 10 * i for i in range (1, 11)]
years = np.arange(1, 11) * 10 + 1980

plt.bar(years + 0.0, daramji, label='daramji', width=2.5, align='center')
plt.bar(years + 2.8, mouse, label='mouse', width=2.5, align='center')
plt.bar(years + 5.6, matbatji, label='matbatji', width=2.5, align='center')

plt.xticks(years, years)
plt.legend()

plt.show()
```
