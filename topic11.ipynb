```python
# gdp 그래프 그리기

import matplotlib.pyplot as plt 
 
# 우리나라의 연간 1인당 국민소득을 각각 years, gdp에 저장
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] 
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3] 
 
# 선 그래프를 그린다. x축에는 years값, y축에는 gdp 값이 표시된다.
plt.plot(years, gdp, color='green', marker='o', linestyle='solid') 
 
# 제목을 설정한다. 
plt.title("GDP per capita") # 1인당 국민소득
 
# y축에 레이블을 붙인다. 
plt.ylabel("dollars") 
plt.savefig("gdp_per_capita.png", dpi=600)  # png 이미지로 저장 가능
plt.show()
```

```python
# 일차함수 그려보기

x = [i for i in range (1, 100)]
y = [i*2 for i in range (1, 100)]

plt.plot (x, y, color='blue') # plot 그리기
plt.title("y = f(x) = 2x") # 제목 붙이기
plt.xlabel("x") # x라벨
plt.ylabel("y") # y라벨
plt.axis([0, 100, 0, 100]) # x의 범위는 0~100, y의 범위는 0~100 잡고 그리기
plt.show()
```

```python
# 한 window에 3개의 함수 그리기
# y = 2x, 3x, 4x

import math

x = [x for x in range (-50, 51)]
y1 = [2*compo for compo in x]
y2 = [3*compo for compo in x]
y3 = [4*compo for compo in x]

# plt.plot(x, y1, 'r-', x, y2, 'g-', x, y3, 'b-')
plt.plot(x, y1, label='2x')
plt.plot(x, y2, label='3x')
plt.plot(x, y3, label='4x')

plt.legend() # 범례 추가 (범례 추가하기 위해서는 label 삽입해야 함)
plt.show()
```

```python
# 사인, 코사인, 탄젠트 

import math

x = list(range(-360, 361))
sine = [math.sin(math.radians(x)) for x in range (-360, 361)]
cosine = [math.cos(math.radians(x)) for x in range (-360, 361)]
tangent = []
for i in x :
    if i % 180 == 90 :
        tangent.append(float('nan'))  # 점근선
    else :
        tangent.append(math.tan(math.radians(i)))

# print (sine, cosine, tangent)

plt.plot(x, sine, label='sine')
plt.plot(x, cosine, label='cosine')
plt.plot(x, tangent, label='tangent')
plt.ylim(-10, 10)  # 매우 큰 tangent값 방지

plt.legend()
plt.show()
```

```python
# 막대형 차트 == plt.bar()
# 점점 늘어나는 다람쥐 마릿수 

years = [1980 + 10 * i for i in range (1, 11)]
daramji = [200 + 2**i for i in range (0, 10)]

plt.bar(years, daramji, width=3)
plt.title('daramji')

plt.xticks (years, years) # x축 눈금

plt.show()
```

```python
# 다중 막대형 그래프

years = [1980 + 10 * i for i in range (1, 11)]
daramji = [200 + 2**i for i in range (0, 10)]
mouse = [80 + 1.5**i for i in range (0, 10)]
matbatji = [100 + 2.1**i for i in range (0, 10)]

plt.bar (years, daramji, label='daramji', width=3)
plt.bar (years, mouse, label='mouse', width=3)
plt.bar (years, matbatji, label='matbatji', width=3)

plt.show()
```

```python
# 위와 같이 하면 알아보기 힘들다 (서로 겹치는 영역)
# 따라서 가로축 범위를 조금씩 차이를 두는 방법을 사용한다 (실수를 더하여)
# ndarray로 변경하여서 계산한다

import numpy as np

# years = [1980 + 10 * i for i in range (1, 11)]
years = np.arange(1, 11) * 10 + 1980

plt.bar(years + 0.0, daramji, label='daramji', width=2.5, align='center')
plt.bar(years + 2.8, mouse, label='mouse', width=2.5, align='center')
plt.bar(years + 5.6, matbatji, label='matbatji', width=2.5, align='center')

plt.xticks(years, years)
plt.legend()

plt.show()
```

```python
# 산포도는 scatter 키워드 사용
# 난수 발생시켜 임의의 2차원 좌표 생성하기

import random

random_x = np.random.randn(10000)
random_y = np.random.randn(10000)

plt.scatter(random_x, random_y, alpha=0.5) # alpha 매개변수에 1보다 작은값 주면 점이 겹칠때 더 진하게 보임
plt.show()
```

```python
# 원형 차트

size = [10000, 5000, 3000, 1000, 1]
sizelabels = ['daramji', 'cat', 'dog', 'mouse', 'lim gyu yeon']

# autopct == 백분율 표시 시 소수점 2번째 자리까지 표시
plt.pie(size, labels = sizelabels, autopct = "%.2f") 
plt.show()
```

```python
# 히스토그램
# bins == 계급의 갯수

randoms = np.random.randn(10000) # 표준정규분포를 따른다
plt.hist(randoms, color='red', bins=50)

randoms = np.random.randn(10000) # 표준정규분포를 따른다
plt.hist(randoms, color='blue', bins=50, alpha=0.5)

plt.show()
```

```python
# 한 화면에 여러 그래프 그리기

data = [np.random.randn(10000), np.random.randn(10000)]
fig, ax = plt.subplots(2, 2, figsize=(5, 5)) # subplots을 선언 후, subplots의 사이즈와 크기를 설정

ax[0, 0].hist(data[1])
ax[0, 1].plot(data[0], data[1], alpha=0.5)
ax[1, 0].scatter(data[0], data[1], alpha=0.5)
ax[1, 1].hist2d(data[0], data[1]) # 2차원 히스토그램
# 각 그래프를 그릴 자리를 ax로 접근

plt.show()
```
