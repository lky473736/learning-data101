```python
# 네 방향으로 내접원 3개 패턴

import turtle

radius = int(turtle.textinput("radius?", "radius?"))

for i in range (4) :
    turtle.circle(radius)
    turtle.circle(radius*2)
    turtle.circle(radius*3)
    turtle.left(90)
```

자료형
- 파이썬은 다른 언어와 다르게 변수 선언 시 자료형을 미리 지정해줄 필요가 없다



- 정수 (integer, fixing point) : -1, 0, 1, 2, 3...
- 실수 (floating point) : -2.1, 1/2, 0.5, 1
- 문자열 (string) : 'daramji'
- 불리언 (boolean) : True, False



- 컴퓨터는 숫자를 위와 같이 IEEE 형식의 실수 저장 (bias) 방식을 사용하기 때문에 오차가 발생할 수밖에 없다
- 따라서 정수 계산이 매우 강한 파이썬에서는 실수를 정수로 변경하여 계산하는 경우도 있음

```python
# 실수의 오차 (정밀도), 동적 형 결정, type

print(0.1 + 0.1 == 0.2)
print(0.1 + 0.1 + 0.1 == 0.3)
print(0.1 + 0.1 + 0.1) # 오차 생김

print (type(1))
print (type('1'))
print (type(bool(1)))
print (type(1.0))
```

```python
# 파이썬의 장점 : 강력한 정수 계산

def frac(n) :
    if n == 0 :
        return 1
    
    else :
        return n * frac(n-1)
    
print ("100! : ", frac(100))
```

```python
# 평균 출력기

sum = 0
cnt = 0

while True : 
    cnt += 1
    compo = input("수 입력 : ")
    
    try : 
        compo = int(compo)
        
    except : 
        break
    
    try : 
        sum += int(compo)
        
    except : 
        break
    
print ("평균 :", sum / cnt)
```
