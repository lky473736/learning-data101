#### 4단원. 조건을 따져 실행해보자



- 순차 구조 : 여러 명령이 순차적으로 진행 (procedure programming)
- 선택 구조 : 여러 개의 명령 중 하나를 선택하여 진행 (if문)
    - 조건문
        - 조건 == boolean expression
        - 만약 조건이 True일 경우 밑의 block이 작동됨
        - indentation을 확실히 해주어야 함 (indentation error가 발생할 수 있기 때문에)
- 반복 구조 : 동일한 명령이 반복

```python
# 조건문 예제 1

speed = int(input('속도를 입력하세요: '))

if speed >= 30 : 
    print ('과속입니다.')
    
else : # 배타적 조건이라고 불림
    print ('안전운전하세요.')
```

```python
# 조건문 예제 2

while True : 
    num = input('정수 입력 : ')
    
    try : 
        num = int(num)
        break
        
    except : 
        print ('정수를 재입력')
        pass
        
if num < 0 : 
    print ('음수')
    
elif num == 0 : 
    print ('0')
    
else : 
    print ('양수')
    
if abs(num) % 2 == 0 :
    print ('짝수')
    
else :
    print ('홀수')
```

```python
# 조건문 예제 3

while True :
    a, b = map(int, input().split())

    if a == b : 
        print ('일치')
        break

    else : 
        print ('불일치')
```

```python
# 조건문 예제 4

import turtle

embedding = {0 : '양수', 1 : '0', 2 : '음수'}
position = [[100, 100 - 100 * i] for i in range(3)]

turtle.penup() # 펜 떨어뜨리기
for i in range (3) : 
    turtle.goto(position[i][0], position[i][1]) # 특정 위치로 거북이를 이동시키는 명령
    turtle.write(f'거북이가 여기로 오면 {embedding[i]}입니다.') # turtle graphic window에 글자를 삽입하는 명령
    
turtle.goto(0, 0) # 제자리 == (0, 0)
turtle.pendown() # 펜 붙이기

while True : 
    mode = turtle.textinput("", "숫자 입력") # GUI에 직접 입력을 넣을 수 있음 (문자열 형태로 반환)
    
    if mode == '00000' : # 사용자가 00000이라 입력하면
        break
    
    try : 
        mode = int(mode) # 형변환 (문자열 -> 정수)

        if mode > 0 : # 양수면
            turtle.goto (position[0][0], position[0][1])

        elif mode < 0 : # 음수면
            turtle.goto (position[2][0], position[2][1])

        else : # 0이면
            turtle.goto (position[1][0], position[1][1])
    
    except :    
        turtle.TK.messagebox.showinfo(title="error", message="다시 입력") # 에러 알림창 출력
        
    turtle.penup()
    turtle.goto(0, 0) # 제자리
    turtle.pendown()

turtle.bye()
```

```python
# 조건문 예제 5

import turtle
turtle.width(3) # 펜 두께
turtle.shapesize(3, 3) # 거북이 사이즈

while True :
    command = input('명령 입력')
    
    match(command) :
        case 'l' : 
            turtle.left(90)
            turtle.forward(100)
            
        case 'r' : 
            turtle.right(90)
            turtle.forward(100)
```

```python
# 조건문 예제 6

year = int(input('해를 입력하세요 : '))

if year % 400 == 0 : 
    print ('윤년')

if year % 4 == 0 and year % 100 != 0 : 
    print (f'{year}년은 윤년임')
    
else : 
    print (f'{year}년은 윤년 아님')
```

```python
# 조건문 예제 7

import random

print ('동전 던지기')
coin = random.choice([0, 1]) # 0과 1 중 아무거나

if coin == 0 : 
    print ('앞면')
    
else : 
    print ('뒷면')
```

```python
# 조건문 예제 8

import random

p1 = input('p1? : ')
p2 = input('p2? : ')

v1 = random.randrange(1, 7) # 1 ~ 6 (7 포함 X, randint라면 포함)
v2 = random.randrange(1, 7) # 1 ~ 6 (7 포함 X, randint라면 포함)

print (v1, v2)

if v1 > v2 : 
    print (f'{p1}이 이김')
    
elif v1 < v2 : 
    print (f'{p2}이 이김')
    
else : 
    print ('비김')
```
