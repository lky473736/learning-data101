```python
# 과속 측정기
# 5번 이상 연속으로 과속 or 10번 이상 과속하면 면허 정지
# 만약 50 ~ 60이면 경고 1회
# 20번 이상 경고 == 정지
# 면허 정지되면 10초 휴식됨, 리셋됨
# 과속 기준은 60

import time 

caution = 0
stop_token = 0
log = []

while True :
    print (log, caution, stop_token)
    if stop_token == 1 : 
        print ('면허를 정지합니다. 10초 동안 차를 운전할 수 없습니다.')
        time.sleep(10)
        
        print ('면허 정지가 해제됩니다. 안전운행하세요.')
        caution = 0
        log = []
        stop_token = 0
        
    speed = int(input("현재 속도를 입력하세요 : "))
    
    if speed <= 0 : 
        print ('운전을 종료합니다.')
        break
    
    if speed >= 60 : 
        print ("과속입니다.")
        log.append ('+')
        
        if len(log) >= 5 and log[-5:] == ['+' for i in range (5)] : # 5번 연속 과속
            stop_token = 1
            continue
            
        if len(log) >= 5 and log.count('+') >= 10 : 
            stop_token = 1
            continue
        
    else : 
        if speed >= 50 and speed < 60 :
            print ("경고합니다.")
            caution += 1
            
            if caution >= 20 :
                stop_token = 1
                continue
            
        log.append ('-')
```

```python
# 홀짝 분류

num = int(input())

if num > 0 and num % 2 == 0 :
    print ("짝수")

elif num > 0 and num % 2 != 0 :
    print ("홀수")
    
else : 
    print ("양수 아님")
```

```python
# 일치할때까지 계속 while

while True : 
    a, b = map(int, input().split())
    
    if a == b : 
        print ("equel")
        break
        
    else : 
        print ("not equel")
```

```python
# 윤년 계산

year = int(input('연도를 입력하세요 : '))

if year % 4 == 0 : 
    if year % 100 == 0 : 
        print ("윤년 아님")
        
    else : 
        print ("윤년임")

elif year % 400 == 0 : 
    print ("윤년임")
```

```python
# 동전 던지기, 주사위 던지기

import random 

mode = int(input("동전 : 0 / 주사위 : 1 ; "))

if mode == 0 : 
    com = random.choice([0, 1])
    user = input("컴퓨터가 동전을 던졌습니다. 앞면? 뒷면? : ")
    
    if user == '앞면' : 
        user = 0
        
    elif user == '뒷면' :
        user = 1
        
    else : 
        print ("오류 발생")
        exit()
        
    if user == com : 
        print ("맞추셨습니다.")
        
elif mode == 1 : 
    com = random.randrange(1, 7)
    user = int(input("컴퓨터가 주사위를 던졌습니다. 1~6? : "))
    
    if com == user : 
        print ("맞추셨습니다.")
```

```python
# user 2명이서 주사위 던지기

import random

user1 = input("첫번째 플레이어의 이름을 입력하세요 : ")
user2 = input("두번째 플레이어의 이름을 입력하세요 : ")

s1 = random.randint(1, 6)
s2 = random.randint(1, 6)

if s1 > s2 : 
    print (user1, '가 이겼습니다.')
    
elif s1 < s2 : 
    print (user2, '가 이겼습니다.')
    
else :
    print (user1, '과', user2, '의 수가 동일합니다.')
```
