```python
#  
# 5    or 10    
#  50 ~ 60  1
# 20   == 
#   10 , 
#   60

import time 

caution = 0
stop_token = 0
log = []

while True :
    print (log, caution, stop_token)
    if stop_token == 1 : 
        print (' . 10     .')
        time.sleep(10)
        
        print ('  . .')
        caution = 0
        log = []
        stop_token = 0
        
    speed = int(input("   : "))
    
    if speed <= 0 : 
        print (' .')
        break
    
    if speed >= 60 : 
        print (".")
        log.append ('+')
        
        if len(log) >= 5 and log[-5:] == ['+' for i in range (5)] : # 5  
            stop_token = 1
            continue
            
        if len(log) >= 5 and log.count('+') >= 10 : 
            stop_token = 1
            continue
        
    else : 
        if speed >= 50 and speed < 60 :
            print (".")
            caution += 1
            
            if caution >= 20 :
                stop_token = 1
                continue
            
        log.append ('-')
```

```python
#  ,  

import random 

mode = int(input(" : 0 /  : 1 ; "))

if mode == 0 : 
    com = random.choice([0, 1])
    user = input("  . ? ? : ")
    
    if user == '' : 
        user = 0
        
    elif user == '' :
        user = 1
        
    else : 
        print (" ")
        exit()
        
    if user == com : 
        print (".")
        
elif mode == 1 : 
    com = random.randrange(1, 7)
    user = int(input("  . 1~6? : "))
    
    if com == user : 
        print (".")
```
