#### 8단원. 연관된 데이터를 딕셔너리로 쌍을 짓자
- 딕셔너리
    - C++에서 hash
    - key-value pair 구조
    - 중괄호, dict()을 통하여 선언함

```python
phone_book = {"daramji" : None, "dedenne" : 112}
print (phone_book["daramji"]) # daramji key에 접근

information = {"daramji" : {"age" : 3, "favorite" : "ice cream"}, 
               "dedenne" : {"age" : 2, "favorite" : "nuts"},
               "pachiritsu" : {"age" : 1, "favorite" : "milk"}} # value엔 아무거나 들어갈 수 있음 (딕셔너리, 튜플, 리스트...)

print (information["daramji"]["age"])
print (information["dedenne"]["favorite"])
```

```python
# 위와 이어서

print (information.keys()) # 딕셔너리의 모든 키들
print (information.values()) # 딕셔너리의 모든 values들
print (information.items()) # 딕셔너리의 각각의 요소가 (key, value) 형태로 반환)

for age, favorite in information.items() :
    print (age, favorite)
```

```python
# 위와 이어서

# 딕셔너리 정렬 (sorted을 이용하여)
items = sorted(information.items(), key = lambda x : x[1]['age']) # value 중 age에 의한 정렬 
print (items)
```

```python
# 위와 이어서

# 딕셔너리의 항목을 삭제하기 위해선 del 사용
information['some'] = 'some'
print (information)

del information['some']
print (information)
```

- set
    - 리스트, 튜플은 index가 존재, dictionary는 순서가 있긴 있지만 key를 부여하여 직접적으로 접근 가능
    - set은 순서가 없음 
        - in, not in을 연산할 때 O(1)
    - set은 중복을 허용하지 않음
    - 중괄호만을 이용하여 사용

```python
seta = {1, 2, 3}
seta.add(4) # 추가
seta.add(1) # 중복 허용 안됨
print (seta)
```

```python
lista = [1, 1, 1, 2, 3, 2, 3, 4, 4, 4]
seta = set(lista)
print (seta) # 중복이 전부 제거되어 있음

sentence = 'daramji is very cute'
seta = set(list(sentence))
print (seta) # 중복 제거
```

```python
# 정렬된 순서로 set을 출력하고 싶다면 sorted를 for문과 엮어서 사용

seta = {2, 1, 3}
for compo in sorted(seta) :
    print (compo)
```

```python
# add, remove

seta = {0, 1, 2, 3, 4, 5}

i = 1

while True :
    if seta == set() :
        print ("모두 삭제 완료")
        break
        
    else :
        if (i % 5) % 2 == 0 :
            seta.remove (i % 5)
        
    i += 1
```

```python
# 집합에서도 min, max, len, stored, sum 다 가능함
# all, any

a_set = {1, 2, 3, 4, 5, 0}
print (all(a_set), any(a_set)) # all : 모두 true인가?, any : 하나라도 true인가?
```

```python
# 합집합, 교집합, 차집합, 부분집합
seta = {1, 2, 3, 4, 5}
setb = {2, 3, 4, 10}

print (seta <= setb) # 부분집합인가
print (setb < seta) # 진부분집합인가

print (seta.union(setb)) # 합집합
print (seta | setb) # 합집합

print (seta.intersection(setb)) # 교집합
print (seta & setb) # 교집합

print (seta.difference(setb)) # 차집합
print (seta - setb) # 차집합 
# 차집합은 교환법칙이 성립하지 않는다

print (seta.symmetric_difference(setb)) # 대칭차집합 (합집합에서 교집합 빼기)
print (seta ^ setb) # 대칭차집합
```

```python
partyA = set(["park", "kim", "lee"])
partyB = set(["park", "choi"])

# 1
print ("A, B에 참석 : ", partyA | partyB)

# 2 
print ("A, B에 중복 없이 : ", partyA ^ partyB)

# 3
print ("A에만 참석", partyA - partyB)
```

- 파일 : 데이터를 저장하기 위해 사용되는 논리적 단위
- 디스크에 저장하여 언제든지 불러올 수 있으며 필요에 따라 수정 가능

```python
# 파일에 쓰기

f = open('data/hello.txt', 'w') # 파일을 w 모드로 연다 (파일 없으면 만들어줌)
f.write ("hello world") # write함
f.close()
```

```python
# 파일 읽기

f = open('data/hello.txt', 'r') # 파일을 r 모드로 연다 (파일 없으면 에러)
content = f.read() # f.read하면 파일을 읽어온다 
print (content)
f.close()

# 파일에 여러 줄이 있다면 for문을 이용해도 됨
```

```python
# 중복되지 않은 단어의 갯수 구하기

def process (w) :
    output = ""
    
    for char in w : 
        if char.isalpha() :
            output += char
            
    return output.lower() # 소문자로 전부 바꾸기 (Well == well)

f = open('data/proverb.txt', 'r') # 파일을 염
# content = f.read() <-- 이거 하면 왜 띄어쓰기 말고 글자를 분리할까...
words = set()

for line in f : 
    wordlist = line.split()
    print (wordlist) # 공백으로 구분된 리스트
    
    for word in wordlist : 
        words.add(process(word))
        
print ("단어 : ", words)
print ("단어 갯수 : ", len(words))
```

```python
# 최대공약수, 최소공배수 구하기
# 유클리드 호제법을 이용

def euclidean (X, Y) :
    a = Y 
    b = X % Y 
    p = 0
    
    if b == 0 :
        return a
    
    while True :
        p = a % b
        
        if p == 0 :
            break
        
        a = b
        b = p
        
    return b

a, b = 10, 15
gcd = euclidean(a, b)
lcm = (a * b) // euclidean(a, b)

print (gcd, lcm)
```

```python
# 집합을 이용하여 계산하기
# 최대공약수 구하기

a, b = 1024, 128
divisors_a = []
divisors_b = []

for i in range (1, a+1) :
    if a % i == 0 : 
        divisors_a.append (i)
        
for i in range (1, b+1) : 
    if b % i == 0 : 
        divisors_b.append (i) 
        
same_divisors = set(divisors_a) & set(divisors_b)
gcd = max(same_divisors)
print (gcd)
```
